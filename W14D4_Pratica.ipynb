Il dataset diabetes.csv raccoglie persone con diabete o meno, e il valore di diverse variabili fisiologiche dei pazienti.
* Osserviamone le dimensioni e un'anteprima di cinque righe;
* Prendiamoci un po' di tempo per dare un'occhiata ai metadati delle colonne;
* Stampiamo dei descrittori statistici del dataset;
* Selezioniamo i dati relativi a diverse fasce di età:
  * <20
  * 20-30
  * 30-40
  * 40-50
  * 50<

* Qual è la media della pressione sanguigna diastolica per le diverse fasce di età?
* Qual è la media della pressione per ogni singolo anno di età?

```python
# importiamo il dataset

diab = pd.read_csv("/content/drive/MyDrive/PYTHON/beginner_datasets/diabetes.csv")
diab
```

```python
# Osserviamone le dimensioni e un'anteprima di cinque righe

print(f"Il dataset contiene {diab.shape[0]} righe e {diab.shape[1]} colonne")
print(f"Il dataset contiene {diab.size} elementi")
print("---")
print("Le prime cinque righe sono:")
diab.head(5)
```

```python
# Prendiamoci un po' di tempo per dare un'occhiata ai metadati delle colonne

print("Gli elementi sono di tipo:")

diab.info()
```

```python
# Stampiamo dei descrittori statistici del dataset

diab.describe()
```

```python
# Selezioniamo i dati relativi a diverse fasce di età: <20, 20-30, 30-40, 40-50, 50<.
# Qual è la media della pressione sanguigna diastolica per le diverse fasce di età?

diab_under20 = diab[diab['Age (years)'] < 20]
diab_20_30 = diab[(diab['Age (years)'] >= 20) & (diab['Age (years)'] < 30)]
diab_30_40 = diab[(diab['Age (years)'] >= 30) & (diab['Age (years)'] < 40)]
diab_40_50 = diab[(diab['Age (years)'] >= 40) & (diab['Age (years)'] < 50)]
diab_over50 = diab[diab['Age (years)'] >= 50]

print("La media della pressione sanguigna diastolica per le persone con età inferiore a 20 anni è:", round(diab_under20['Diastolic blood pressure (mm Hg)'].mean(), 2))
print("La media della pressione sanguigna diastolica per le persone con età compresa tra 20 e 30 anni è:", round(diab_20_30['Diastolic blood pressure (mm Hg)'].mean(), 2))
print("La media della pressione sanguigna diastolica per le persone con età compresa tra 30 e 40 anni è:", round(diab_30_40['Diastolic blood pressure (mm Hg)'].mean(), 2))
print("La media della pressione sanguigna diastolica per le persone con età compresa tra 40 e 50 anni è:", round(diab_40_50['Diastolic blood pressure (mm Hg)'].mean(), 2))
print("La media della pressione sanguigna diastolica per le persone con età superiore a 50 anni è:", round(diab_30_40['Diastolic blood pressure (mm Hg)'].mean(), 2))
```

```python
# Qual è la media della pressione per ogni singolo anno di età?

print("La media della pressione sanguigna diastolica per ogni singolo anno di età è: \n")
round(diab.groupby('Age (years)')['Diastolic blood pressure (mm Hg)'].mean(), 2)
```

Il dataset insurance.csv contiene dati rispetto a caratteristiche e abitudini delle persone, e della zona in cui vivono, rispetto ai costi individuali per le cure mediche come premio per le assicurazioni sulla salute.
* Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna
* Quali sono le medie di charges rispetto a region? Ci sono differenze significative?
* E rispetto a smoker? E a sex?
* Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?

```python
# importiamo il dataset

ins = pd.read_csv("/content/drive/MyDrive/PYTHON/beginner_datasets/insurance.csv")
ins
```

```python
# Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna

print(f"il dataset contiene {ins.shape[0]} righe e {ins.shape[1]} colonne")
print(f"il dataset contiene {ins.size} elementi")
print("---")
print("I nomi delle colonne sono:")
for i in ins.columns:
  print(" -", i)
print("---")
print("Questa è un anteprima del dataset: \n")
ins
```

```python
# Quali sono le medie di charges rispetto a region? Ci sono differenze significative?

print("Le medie di charges rispetto a region sono: \n")
round(ins.groupby('region')['charges'].mean(), 2)
```

```python
# E rispetto a smoker?

print("Le medie di charges rispetto a smoker sono: \n")
round(ins.groupby('smoker')['charges'].mean(), 2)
```

```python
#E a sex?

print("Le medie di charges rispetto a sex sono: \n")
round(ins.groupby('sex')['charges'].mean(), 2)
```

```python
# Quali sono i descrittori statistici di bmi?

print("I descrittori statistici di bmi sono: \n")
ins.bmi.describe()
```

```python
# Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?

primo_quartile = ins[ins['bmi'] <= ins['bmi'].quantile(0.25)]
print("PRIMO QUARTILE:")
print("La media dei valori di charges rispetto al primo quartile dei valori di bmi è:", round(primo_quartile['charges'].mean(), 2))
print("Il valore minimo di charges rispetto al primo quartile dei valori di bmi è:", primo_quartile['charges'].min())
print("Il valore massimo di charges rispetto al primo quartile dei valori di bmi è:", primo_quartile['charges'].max())
print("---")
secondo_quartile = ins[(ins['bmi'] > ins['bmi'].quantile(0.25)) & (ins['bmi'] <= ins['bmi'].quantile(0.5))]
print("SECONDO QUARTILE:")
print("La media dei valori di charges rispetto al secondo quartile dei valori di bmi è:", round(secondo_quartile['charges'].mean(), 2))
print("Il valore minimo di charges rispetto al secondo quartile dei valori di bmi è:", secondo_quartile['charges'].min())
print("Il valore massimo di charges rispetto al secondo quartile dei valori di bmi è:", secondo_quartile['charges'].max())
print("---")
terzo_quartile = ins[(ins['bmi'] > ins['bmi'].quantile(0.5)) & (ins['bmi'] <= ins['bmi'].quantile(0.75))]
print("TERZO QUARTILE:")
print("La media dei valori di charges rispetto al terzo quartile dei valori di bmi è:", round(terzo_quartile['charges'].mean(), 2))
print("Il valore minimo di charges rispetto al terzo quartile dei valori di bmi è:", terzo_quartile['charges'].min())
print("Il valore massimo di charges rispetto al terzo quartile dei valori di bmi è:", terzo_quartile['charges'].max())
print("---")
quarto_quartile = ins[ins['bmi'] > ins['bmi'].quantile(0.75)]
print("QUARTO QUARTILE:")
print("La media dei valori di charges rispetto al quarto quartile dei valori di bmi è:", round(quarto_quartile['charges'].mean(), 2))
print("Il valore minimo di charges rispetto al quarto quartile dei valori di bmi è:", quarto_quartile['charges'].min())
print("Il valore massimo di charges rispetto al quarto quartile dei valori di bmi è:", quarto_quartile['charges'].max())
```
