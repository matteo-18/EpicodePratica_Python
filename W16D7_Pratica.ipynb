Sul database titanic, effettuiamo le seguenti analisi:
* Quanti ponti c'erano sulla nave?
* Visualizzare un grafico con il numero di passeggeri di ogni classe di imbarco
* Fare la stessa cosa per la colonna alive
* Qual era la distribuzione delle tariffe (fare)?

```python
# Quanti ponti c'erano sulla nave?

print(f"Sulla nave, secondo i dati disponibili nel dataset, erano presenti {tit['Cabin'].str[0].nunique()} ponti")
print("---")
print("Numero di cabine per ogni ponte secondo i dati disponibili nel dataset:")
tit['Cabin'].str[0].value_counts()
```

```python
# Visualizzare un grafico con il numero di passeggeri di ogni classe di imbarco

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
sns.countplot(x ='PClass', hue = 'PClass',  data=tit, palette = 'rainbow')
```

```python
# Fare la stessa cosa per la colonna alive

sns.countplot(x = 'Survived', hue = 'Survived',  data = tit, palette = 'rainbow')
```

```python
# Qual era la distribuzione delle tariffe (fare)?

sns.displot(tit['Fare'])
```

```python
# Riusciamo a vedere la distribuzione delle età dei passeggeri rispetto alla classe di imbarco? Proviamo con un boxplot e con un violinplot

sns.boxplot(x = 'PClass', y = 'Age', data = tit)
```

```python
sns.violinplot(x = 'PClass', y = 'Age', data = tit)
```

```python
# Visualizziamo un boxplot rispetto alle colonne fare e survived; che cose ne deduciamo?

sns.boxplot(x = 'Survived', y = 'Fare', data = tit)
```

Scarichiamo il dataset stockdata.csv da https://github.com/plotly/datasets/blob/master/stockdata.csv e:
* Visualizziamo l'andamento delle azioni della Apple ("AAPL")
* Visualizziamo l'andamento di tutte le azioni del dataset
* Visualizziamo l'andamento di Microsoft ("MSFT") e IBM ("IBM") negli anni 2012 e 2013
* Con un istogramma, troviamo il range di valori che le azioni di Microsoft hanno assunto più spesso

```python
# importiamo il daset

stock = pd.read_csv("/content/drive/MyDrive/PYTHON/stockdata (1).csv")
stock
```

```python
# Visualizziamo l'andamento negli anni delle azioni della Apple ("AAPL"). Sull'asse delle x devono esserci solo gli anni

stock['Date'] = pd.to_datetime(stock['Date'])
sns.lineplot(x = 'Date' , y = 'AAPL', data = stock)
```

```python
# Visualizziamo l'andamento di tutte le azioni del dataset

stock.plot(x = 'Date')
```

```python
# Visualizziamo l'andamento di Microsoft ("MSFT") e IBM ("IBM") negli anni 2012 e 2013

stock[stock['Date'].dt.year.isin([2012, 2013])].plot(x = 'Date', y = 'IBM')
stock[stock['Date'].dt.year.isin([2012, 2013])].plot(x = 'Date', y = 'MSFT')
stock[stock['Date'].dt.year.isin([2012, 2013])].plot(x = 'Date', y = ['MSFT', 'IBM'])
```

```python
# Con un istogramma, troviamo il range di valori che le azioni di Microsoft hanno assunto più spesso

sns.histplot(stock['MSFT'])
```

Scarichiamo il dataset metal_bands_2017.csv da https://www.kaggle.com/datasets/mrpantherson/metal-by-nation e:
* con Seaborn visualizziamo l'istogramma del numero dei fan delle varie band; cosa si vede? Cose ne deduciamo?
* Delle band che suonano sia Progressive che Thrash, realizziamo un barplot che mostri quanti fan ci sono in totale rispetto al paese di origine (nota: in questo caso specifico, per accedere alla colonna "style" dovremo usare ["style"] e non .style perché quest'ultimo è un metodo
* Visualizziamo un istogramma delle date di formazione delle band (potremo dover convertire in qualche modo questo dato); si nota qualche trend?

```python
# importiamo il dataset

metal = pd.read_csv("/content/drive/MyDrive/PYTHON/metal_bands_2017 (1).csv", encoding = 'latin1')
metal
```

```python
# con Seaborn visualizziamo l'istogramma del numero dei fan delle varie band; cosa si vede? Cose ne deduciamo?

sns.histplot(x = metal['fans'], bins = 100)
```

```python
# Delle band che suonano sia Progressive che Thrash, realizziamo un barplot che mostri quanti fan ci sono in totale rispetto al paese di origine
# (nota: in questo caso specifico, per accedere alla colonna "style" dovremo usare ["style"] e non .style perché quest'ultimo è un metodo

prog_trash = metal[(metal['style'] == 'Progressive') | (metal['style'] == 'Thrash')]
plt.figure(figsize = (5, 10))
sns.barplot(x = 'fans', y = 'origin', data = prog_trash)
```

```python
# Visualizziamo un istogramma delle date di formazione delle band (potremo dover convertire in qualche modo questo dato); si nota qualche trend?

import numpy as np

metal['formed'] = metal['formed'].replace('-', np.nan) #sostituiamo tutti gli elementi che hanno il "-" nella colonna "origin" con valori nulli
metal['formed'] = pd.to_datetime(metal['formed']) #convertiamo in datetime
sns.histplot(metal['formed'])
```

Scarichiamo il dataset elections.csv da https://github.com/plotly/datasets/blob/master/election.csv:
* Con un grafico a barre confrontiamo i voti totali presi dai tre candidati (come somma di tutti i distretti)
* Con un grafico a barre confrontiamo il numero di votanti per ogni distretto
* Visualizzare un grafico a barre comparativo dove si confrontano i voti presi nei primi 4 distretti per ogni candidato

```python
# importiamo il dataset

elec = pd.read_csv("/content/drive/MyDrive/PYTHON/election.csv")
elec
```

```python
# Con un grafico a barre confrontiamo i voti totali presi dai tre candidati (come somma di tutti i distretti)

voti_tot = {'Coderre' : elec.Coderre.sum(), 'Bergeron' : elec.Bergeron.sum(), 'Joly' : elec.Joly.sum() }
voti_tot = pd.Series(voti_tot)

sns.barplot(x = voti_tot.index, y = voti_tot.values, hue = voti_tot.index, palette ='pastel')
```

```python
# Con un grafico a barre confrontiamo il numero di votanti per ogni distretto

plt.figure(figsize = (25, 5))
sns.barplot(x = 'district_id', hue = 'district_id' , y = 'total' , data = elec , palette = 'rainbow', legend = False)
```

```python
# Visualizzare un grafico a barre comparativo dove si confrontano i voti presi nei primi 4 distretti per ogni candidato

voti_top4 = {'Coderre' : elec.nlargest(4, 'Coderre')['Coderre'].sum(), 'Bergeron' : elec.nlargest(4, 'Bergeron')['Bergeron'].sum(), 'Joly' : elec.nlargest(4, 'Joly')['Joly'].sum()}
voti_top4 = pd.Series(voti_top4)

sns.barplot(x = voti_top4.index, y = voti_top4.values, hue = voti_top4.index, palette ='pastel')
```

```python
elec.head(4)[['Coderre', 'Bergeron', 'Joly', 'district_id']].set_index('district_id').plot(kind = 'bar')
```

```python
elec.nsmallest(4, 'district_id')[['Coderre', 'Bergeron', 'Joly', 'district_id']].set_index('district_id').plot(kind = 'bar')
```

Scarichiamo i file party_in_nyc.csv dal dataset https://www.kaggle.com/datasets/somesnm/partynyc relativo a quante volte è stata chiamata la polizia per delle feste eccessivamente rumorose:
* Realizziamo un grafico che mostra quante chiamate ci sono state per ogni città per tutti i tipi di locale
* Qual è il rettangolo di latitudine e longitudine considerato dal dataset?
* Quante segnalazioni sono state fatte per case private?
* Con Seaborn visualizziamo il conteggio delle segnalazioni avvenute in ogni quartiere; che succede se aggiungiamo il parametro hue="Location Type"?

```python
# importiamo il dataset

party = pd.read_csv("/content/drive/MyDrive/PYTHON/party_in_nyc.csv")
party
```

```python
# Realizziamo un grafico che mostra quante chiamate ci sono state per ogni città per tutti i tipi di locale

plt.figure(figsize = (5, 15))
sns.countplot(y = 'City', data = party, order = party['City'].value_counts().index, palette = 'rainbow')
```

```python
# Qual è il rettangolo di latitudine e longitudine considerato dal dataset?

party[['Latitude', 'Longitude']].describe().loc[['min', 'max']].transpose()
```

```python
# Quante segnalazioni sono state fatte per case private?

party['Location Type'].value_counts()
```

```python
# Con Seaborn visualizziamo il conteggio delle segnalazioni avvenute in ogni quartiere

sns.countplot(x = 'Borough', data = party)
```

```python
# che succede se aggiungiamo il parametro hue="Location Type"?

plt.figure(figsize = (10, 10))
sns.countplot(x = 'Borough', data = party, hue = 'Location Type', palette = 'pastel')
```
