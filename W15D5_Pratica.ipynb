Nei beginner_datasets, il dataset wine.csv contiene delle analisi organolettiche su diverse qualit√† di vini:
* Qual √® la media di concentrazione alcolica per ogni qualit√†? Ci sono differenze? E rispetto alla media totale?
* C'√® differenza nella concentrazione alcolica per vini bianchi e vini rossi?
* Rifacendo le analisi delle domande precedenti ma per il pH, cambia qualcosa?
* E per i solfati?

```python
#importiamo il dataset

wine = pd.read_csv("/content/drive/MyDrive/PYTHON/beginner_datasets/wine.csv")
wine
```

```python
# Qual √® la media di concentrazione alcolica per ogni qualit√†? Ci sono differenze? E rispetto alla media totale?

print("La media di concentrazione alcolica del totale del dataset √®: ", round(wine['alcohol'].mean(), 2))
print("---")
print("La media di concentrazione alcolica per ogni qualit√† √®: \n")
round(wine.groupby('quality')['alcohol'].mean(), 2)
```

```python
# C'√® differenza nella concentrazione alcolica per vini bianchi e vini rossi?

print("La media di concentrazione alcolica dei vini bianchi √®: ", round(wine[wine['type'] == 'white']['alcohol'].mean(), 2))
print("La media di concentrazione alcolica dei vini rossi √®: ", round(wine[wine['type'] == 'red']['alcohol'].mean(), 2))
```

```python
# Rifacendo le analisi delle domande precedenti ma per il pH, cambia qualcosa?

print("La media del valore del pH del totale del dataset √®: ", round(wine['pH'].mean(), 2))
print("---")
print("La media del valore del pH dei vini bianchi √®: ", round(wine[wine['type'] == 'white']['pH'].mean(), 2))
print("La media del valore del pH dei vini rossi √®: ", round(wine[wine['type'] == 'red']['pH'].mean(), 2))
print("---")
print("La media del valore del pH per ogni qualit√† √®: \n")
round(wine.groupby('quality')['pH'].mean(), 2)
```

```python
# E per i solfati?

print("La media del valore dei sulfati del totale del dataset √®: ", round(wine['sulphates'].mean(), 2))
print("---")
print("La media del valore dei sulfati dei vini bianchi √®: ", round(wine[wine['type'] == 'white']['sulphates'].mean(), 2))
print("La media del valore dei sulfati dei vini rossi √®: ", round(wine[wine['type'] == 'red']['sulphates'].mean(), 2))
print("---")
print("La media del valore dei sulfati per ogni qualit√† √®: \n")
round(wine.groupby('quality')['sulphates'].mean(), 2)
```

Nei beginner_datasets, il file boston.csv contiene il Boston Housing Dataset, che deriva dalle informazioni raccolte dal Census Service degli Stati Uniti sulle abitazioni nell'area di Boston. Di seguito vengono descritte le colonne del dataset:
* CRIM - tasso di criminalit√† pro capite per citt√†
* ZN - proporzione di terreni residenziali suddivisi in zone per lotti superiori a 25.000 piedi quadrati.
* INDUS - percentuale di acri di attivit√† commerciali non al dettaglio per citt√†.
* CHAS - variabile dummy del fiume Charles (1 se il tratto costeggia il fiume; 0 altrimenti)
* NOX - concentrazione di ossidi di azoto (parti per 10 milioni)
* RM - numero medio di stanze per abitazione
* AGE - proporzione di unit√† abitative occupate da proprietari costruite prima del 1940
* DIS - distanze ponderate da cinque centri occupazionali di Boston
* RAD - indice di accessibilit√† alle autostrade radiali
* TAX - aliquota dell'imposta fondiaria sul valore pieno per 10.000 dollari
* PTRATIO - rapporto alunni-insegnanti per citt√†
* BLACK - la percentuale di neri per citt√†
* LSTAT - % di popolazione di condizione pi√π bassa
* MEDV - Valore mediano delle case, espresso in migliaia di dollari

Esercizio:
* La media del prezzo delle case cambia a seconda della distanza dal fiume Charles?
* Si nota una correlazione tra il tasso di criminalit√† e il valore delle abitazioni? Come si pu√≤ spiegare il risultato?
* Qual √® la media del numero di stanze rispetto al rapporto alunni-insegnanti? E del valore delle case? Appare esserci una qualche correlazione? Come si pu√≤ spiegare il risultato?
* Rispetto all'accessibilit√† alle autostrade, cambia qualcosa la media delle et√† delle abitazioni? E del numero di stanze? E delle tasse?

```python
# importiamo il dataset

bos = pd.read_csv("/content/drive/MyDrive/PYTHON/beginner_datasets/boston.csv")
bos
```

```python
# La media del prezzo delle case cambia a seconda della distanza dal fiume Charles?

print(f"La media del prezzo delle case che costeggiano il fiume √®: {round((bos[bos['chas'] == 1]['medv'].mean() * 1000) , 2)} $")
print(f"La media del prezzo delle case che non costeggiano il fiume √®: {round((bos[bos['chas'] == 0]['medv'].mean() * 1000) , 2)} $")

bos.groupby('chas')['medv'].mean()

# Osserviamo che le case costruite nelle zone che costeggiano il fiume hanno un valore medio pi√π alto di pi√π di 6000 $
```

```python
# Si nota una correlazione tra il tasso di criminalit√† e il valore delle abitazioni? Come si pu√≤ spiegare il risultato?

bos.groupby(pd.qcut(bos['crim'], 3, labels = ['Basso tasso di criminalit√†', 'Medio tasso di criminalit√†', 'Alto tasso di criminalit√†']), observed = True)['medv'].mean() * 1000

# Osserviamo che il valore medio delle abitazioni √® inversamente proporzionale al valore del tasso di criminalit√†.
# Ci√≤ significa che nelle zone pi√π sicure e con minor tasso di criminalit√†, il prezzo medio delle abitazioni aumenta.
```

```python
# Qual √® la media del numero di stanze rispetto al rapporto alunni-insegnanti?
print("La media del numero di stanze rispetto al rapporto alunni-insegnanti √®: \n")
bos.groupby(pd.qcut(bos['ptratio'], 3, labels = ['Primo Terzile' , 'Secondo Terzile' , 'Terzo Terzile']), observed = True)['rm'].mean()

# Osserviamo che per bassi rapporti alunni-insegnanti (ovvero minor numero di studenti per insegnante), il numero delle stanze medio delle abitazioni √® pi√π alto.
# Per alti rapporti alunni-insegnanti (ovvero maggior numero di studenti per insegnante), il numero delle stanze medio delle abitazioni √® pi√π basso.
# Nelle zone con maggior numero di studenti per insegnante, le case hanno meno stanze quindi, presumibilmente, valgono di meno
```

```python
# E del valore delle case? Appare esserci una qualche correlazione? Come si pu√≤ spiegare il risultato?

print("La media del valore delle case rispetto al rapporto alunni-insegnanti √®: \n")
bos.groupby(pd.qcut(bos['ptratio'], 3, labels = ['Primo Terzile' , 'Secondo Terzile' , 'Terzo Terzile']), observed = True)['medv'].mean()

# Confermiamo quanto ipotizzato, nell'analisi precedente, il valore medio delle case √® minore nelle zone con maggior numero di studenti per insegnante.
```

```python
# Rispetto all'accessibilit√† alle autostrade, cambia qualcosa la media delle et√† delle abitazioni?

bos.groupby(pd.qcut(bos['rad'], 3), observed = True)['age'].mean()

# Osserviamo che per le zone con pi√π alto indice di accessibilit√† alle autostrade radiali, la % delle abitazioni costruite prima del 1940 √® pi√π alta
```

```python
# E del numero di stanze?

bos.groupby(pd.qcut(bos['rad'], 3), observed = True)['rm'].mean()

# Osserviamo che per le zone con pi√π alto indice di accessibilit√† alle autostrade radiali, il numero medio di stanze √® lievemente inferiore rispetto alle zone con indice pi√π basso.
```

```python
# E delle tasse?

bos.groupby(pd.qcut(bos['rad'], 3), observed = True)['tax'].mean()

# Osserviamo che per le zone con pi√π alto indice di accessibilit√† alle autostrade radiali, le tasse sono maggiori rispetto alle zone con indice pi√π basso.
```
